```python
# importação das bibliotecas
import smtplib 
import email.message

# método eviar_email
def enviar_email(loja, resumo_loja):
  # criando o conteúdo da menssagem
  corpo_email = f'''
  <p style="text-align: justfy;">Caríssimos ao tempo que lhe cumprimentamos cordialmente,enviamos o relatório gerencial da loja sob sua responabilidade</p>
  {resumo_loja.to_html()}
  <p style="text-align: justfy;">Sem mais para o momento, nos despedimos e nos colocamos a sua inteira disposição para quaisquer esclarecimentos.</p>
  <p style="text-align: justfy;">Atenciosamente, equipe de sistemas.</p>
  '''

  # criando o email
  msg = email.message.Message()
  msg['Sunject'] = f'Relatório Gerencial - {loja}'
  # conta que será usada para o envio
  msg['From'] = 'remetentte@gmail.com'
  # senha da conta que será usada para o envio
  password = 'senha'
  # conta para o qual o email será enviado (não precisa ser gmail.com)
  msg['To'] = 'destinatario@gmail.com'
  msg.add_header('Content-Type, text/html')
  msg.set_payload(corpo_email)

  # estabelecendo uma conexão com Servidor SMTP
  server = smtplib.SMTP('smtp.gmail.com:587')

  # iniciando uma conexão segura TLS (Transport Layer Security) com Servidor SMTP
  server.starttls()

  # passagem de credenciais para enviar o e-mail
  server.login(msg['From'], password)
  # envio da mensagem
  server.sendmail(msg['From'], [msg['To']], msg.as_string().encode('utf-8'))

  print('E-mail enviado')
```

```python
loja = 'Bananeiras/PB'
# Filtrando os dados para uma filial definida pela loja
df_filial = df.loc[df['ID Loja'] == loja, ['ID Loja', 'Quantidade', 'Valor Final']]

# Agrupando e cauculando o somatório dos registros
relatorio_filial = df_filial.groupby('ID Loja').sum()

# Cauculando o valor médio das vendas do periodo
relatorio_filial['Ticket Médio'] = relatorio_filial['Valor Final']/relatorio_filial['Quantidade']

display(relatorio_filial)
```

```python
# criando a lista de filiais
lojas = df['ID Loja'].unique()

for loja in lojas:
  # Filtrando os dados para uma filial definida pela loja
  df_filial = df.loc[df['ID Loja'] == loja, ['ID Loja', 'Quantidade', 'Valor Final']]

  # Agrupando e cauculando o somatório dos registros
  relatorio_filial = df_filial.groupby('ID Loja').sum()

  # Cauculando o valor médio das vendas do periodo
  relatorio_filial['Ticket Médio'] = relatorio_filial['Valor Final']/relatorio_filial['Quantidade']

  display(relatorio_filial)
```
